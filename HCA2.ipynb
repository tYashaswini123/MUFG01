```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# Step 1: Define the data
data = np.array([
    [25, 40],  # Customer 1
    [34, 65],  # Customer 2
    [22, 50],  # Customer 3
    [45, 80],  # Customer 4
    [29, 60],  # Customer 5
    [40, 70]   # Customer 6
])

labels = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6']

# Step 2: Perform single linkage hierarchical clustering
linkage_matrix = linkage(data, method='single', metric='euclidean')

# Step 3: Plot the dendrogram with distance annotations
plt.figure(figsize=(10, 6))
ddata = dendrogram(linkage_matrix, labels=labels)

# Annotate merge distances
for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
    x = 0.5 * (i[1] + i[2])
    y = d[1]
    plt.plot(x, y, 'o', c=c)
    plt.text(x, y + 1, f"{y:.2f}", ha='center', va='bottom', fontsize=9, color=c)

plt.title("Hierarchical Clustering Dendrogram (Customer Age & Income)")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.grid(True)
plt.tight_layout()
plt.show()
```
