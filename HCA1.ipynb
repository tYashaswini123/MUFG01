```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
import numpy as np

# Define the data points
points = np.array([
    [0.40, 0.53],  # p1
    [0.22, 0.38],  # p2
    [0.35, 0.32],  # p3
    [0.26, 0.19],  # p4
    [0.08, 0.41],  # p5
    [0.45, 0.30]   # p6
])

# Optional: Labels for the points
labels = ['p1', 'p2', 'p3', 'p4', 'p5', 'p6']

# Create the dendrogram using linkage
linked = sch.linkage(points, method='single', metric='euclidean')

# Plot the dendrogram
plt.figure(figsize=(8, 5))
dendrogram = sch.dendrogram(linked, labels=labels, distance_sort='ascending')
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.grid(True)
plt.tight_layout()
plt.show()
```
